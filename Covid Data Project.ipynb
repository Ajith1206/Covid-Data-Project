# # Importing Libraries

```python
import pandas as pd
import numpy as np
import datetime
import re
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("covid_19_india.csv")  # uploading File
print(df.isna().sum())     # checking is any null value
print('\n')
print(df.info())  # summary of datatypes
```

# Data Cleaning

# States wise Total No of Confirmed Cases

```python
df=pd.read_csv("covid_19_india.csv")
                                                                                # replacing values in original file
    
df[df["State/UnionTerritory"]=="Dadra and Nagar Haveli"]
df.replace("Dadra and Nagar Haveli","Dadra and Nagar Haveli and Daman and Diu",inplace=True)
df.replace("Telangana","Telengana",inplace=True)
df.replace("Bihar****","Bihar",inplace=True)
df.replace("Daman & Diu","Dadra and Nagar Haveli and Daman and Diu",inplace=True)
df.replace("Unassigned","Unknown State",inplace=True)
df.replace("Cases being reassigned to states","Unknown State",inplace=True)
                                                                                   # grouping by State wise
group=df.groupby("State/UnionTerritory")
updated_group=group["Confirmed"].agg(["sum"]).sort_values(by="sum",ascending=False)    # sorting
Y_axis=(updated_group["sum"])
X_axis=(updated_group.index)
x_axis=pd.Series(X_axis)
print(Y_axis)
                                                                              # Creating graph
plt.figure(figsize=(20,10))
plt.bar(X_axis,Y_axis,color="turquoise")
plt.title("States wise Total No of Confirmed Cases ",fontsize=30)
plt.xticks(fontsize=20)
plt.ylabel("Total",fontsize=25)
plt.xlabel("States",fontsize=25)
plt.yticks(fontsize=22)
plt.xticks(rotation=40,horizontalalignment="right",fontsize=20)
plt.tight_layout()
plt.show()
```

# States wise Total No of Deaths Cases


```python
group=df.groupby("State/UnionTerritory")   # grouping by State wise
updated_group=group["Deaths"].agg(["sum"]).sort_values(by="State/UnionTerritory",ascending=False) # sorting
Y_axis=(updated_group["sum"])
X_axis=(updated_group.index)
X_axis=pd.Series(X_axis)
print(Y_axis)
                                                        # Creating graph
plt.figure(figsize=(24,14))
plt.bar(X_axis,Y_axis,color="red")
plt.title("States wise Total No of Deaths Cases ",fontsize=30)
plt.xticks(fontsize=20)
plt.ylabel("Total",fontsize=25)
plt.xlabel("States",fontsize=25)
plt.yticks(fontsize=22)
plt.xticks(rotation=40,horizontalalignment="right",fontsize=20)
plt.tight_layout()
plt.show()
```

# Datails of Unknown States

```python
df2=df[df["State/UnionTerritory"]=="Unknown State"]     #df2 inserted all unknown states data
df2["Date"]=pd.to_datetime(df2["Date"])
df2["Years"]=df2["Date"].dt.year                 #formatting in date time format 
df2["Months"]=df2["Date"].dt.month              #extract data in year , months and days
df2["Days"]=df2["Date"].dt.day
df2["ConfirmedIndianNational"].replace("-",int(0),inplace=True)    #replacing(-) into int(0) value
df2["ConfirmedForeignNational"].replace("-",int(0),inplace=True)      #replacing(-) into int(0) value
```

# Adding New Column as Daily Cases

```python
convert=df2["Confirmed"].tolist()                      # converting Confirmed value to list
diff_list=[]                                               #  created empty list
copied_list=convert.copy()                                              #copied list into new variable
index_convert=convert[0]     
copied_list.pop(0)                                             #removing index[0] value
copied_list.append(int(0))
for x in range(0,len(convert)):
    diff=copied_list[x]-convert[x]  
    diff_list.append(diff)                          
diff_list.insert(0,index_convert)                        # inserting covert[0] value in diff_list
diff_list.pop()                                       # removeing the last value 
diff_list
df2["Daily_Cases"]=diff_list                      # adding column as Daily_Cases
df2
```

# Day wise Daily Cases

```python
def funct():                                     # created function
    mon=int(input("Enter the Month"))                                        # takes input
    updated_group=Greater_above_Zero[Greater_above_Zero["Months"]==mon] #comparing
    case=updated_group[["Daily_Cases","Days"]] # grouping and # sorting
    X_axis=case["Days"]
    Y_axis=case["Daily_Cases"]
                                                           # Creating graph                    
    plt.figure(figsize=(15,5))
    plt.bar(X_axis,Y_axis,color="darkorange")
    plt.xlabel("Days",fontsize=15)
    plt.ylabel("Daily_Cases",fontsize=15)
    plt.title("Daily Cases of Day wise  ",fontsize=20)
    plt.tight_layout()
    print(case)
    return case.count()
    
    
funct()
```

# Single state of year 2020 cases of Death, confirmed and Cured


```python

def State():                                                #created function
    state=input("Entered State")
    year=int(input("Entered Year "))      
    data=df1[df1["State/UnionTerritory"]==state] 
    updated_data=data[data["Years"]==year]
    By_Month_Base=updated_data.groupby("Months").agg(["sum"])           # grouping and # sorting
    dead=By_Month_Base["Deaths"]
    death_line=dead["sum"]
    cure=By_Month_Base["Cured"]
    cured_line=cure["sum"]
    conf=By_Month_Base["Confirmed"]
    confirmed_line=conf["sum"]
    month=By_Month_Base.index.sort_values()
                                                       #creating graph
    plt.figure(figsize=(12,8))
    plt.title("Month wise Confirmed , Cured and Deaths Cases ",fontsize=18)
    plt.xlabel("Months",fontsize=18)
    plt.ylabel("Sum value",fontsize=18)
    plt.xticks(fontsize=22)
    plt.yticks(fontsize=22)
    plt.legend(["Confirmed","Cured","Deaths"],loc="upper left",fontsize=18)
    plt.plot(month,death_line,"ro",linewidth=4)
    plt.plot(month,cured_line,"go")
    plt.plot(month,confirmed_line,"b^")
    plt.legend(["Deaths","Cured","Confirmed"],loc="upper left",fontsize=18)
    plt.tight_layout()
State()
```

# # Pie Chart of Percentage of Daily_Cases , Cured and Deaths by day

```python
state=input("Enter the state ")                   #input 
year=int(input("Enter the year "))
month=int(input("Enter the month number "))
date=int(input("Enter the date"))
                                                     # pulling data
updated=df1[(df1["State/UnionTerritory"]==state)&(df1["Years"]==year) & (df1["Months"]==month) & (df1["Days"]==date)]
                                            
Category=["Cured","Deaths","Daily_Cases"]
Quantity=[]                                          #empty list
cured_mean=updated["Cured"].mean()
Quantity.append(cured_mean)               

death_mean=updated["Deaths"].mean()
Quantity.append(death_mean)

day=updated["Daily_Cases"].mean()
Quantity.append(day)
                                                    #creating graph
plt.figure(figsize=(33,8))
my_explode=[0.1,0.1,0.2]
plt.pie(Quantity,labels=Category,shadow=True,autopct='%1.2f%%',
        explode=my_explode,wedgeprops={"edgecolor":"k"})
plt.title("Percentage of Daily_Cases , Cured and Deaths by day",fontsize=20)
plt.tight_layout()
plt.show()
```
